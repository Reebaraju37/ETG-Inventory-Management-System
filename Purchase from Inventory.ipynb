```python
import json

fp = open("record.json","r")
r = fp.read()
fp.close()

records = json.loads(r)
```

```python
records
```

```python
import time 

print("******************---WELCOME---****************\n")

ui_prdid = input("Enter the product id : ")
ui_prqt = int(input("Enter the Quantity of product : "))

if(ui_prqt > records[ui_prdid]['qn']):
    print("Available quntity of ",records[ui_prdid]['name'], " is ", records[ui_prdid]['qn'])
else:
    print("Name of the product: ",records[ui_prdid]['name'])
    print("Price of the product: Rs.",records[ui_prdid]['pr'])
    print("*******************************************")
    print("Billing Amount: Rs.",records[ui_prdid]['pr'] * ui_prqt)
    print("*******************************************")
    print("-------------------------------------------")
    shop_time = time.ctime()
    print("\nTime of shopping :",shop_time )
    print("-------------------------------------------")
    print("**********THANKYOU FOR SHOPPING************")
   
    
    records[ui_prdid]['qn'] = records[ui_prdid]['qn'] - ui_prqt 


js = json.dumps(records)

fp = open("record.json","w")
fp.write(js)
fp.close()


#updating transaction in sales.json file

fr = open("sales.json","r")
n = fr.read()
fr.close()

tsale = json.loads(n)

tsale[(len(tsale)+1)] =  {'prod_id' : ui_prdid,'name':records[ui_prdid]['name'], 'qn': ui_prqt, 'price':records[ui_prdid]['pr'], 'bill_amount' : records[ui_prdid]['pr'] * ui_prqt, 'time': shop_time}

sf = json.dumps(tsale)

fd = open("sales.json","w")
fd.write(sf)
fd.close()
```
